Question 7. From the tests of normality that I ran below, it appears that the assumption of normality is met. We can see that our p value numerically is less than 0.05 and that we have a large f value to accompany it. That showcases that assumption numerically. Then, if we were to review our residual scatterplot, we can see that our data fits on our line relatively normally. There are a few values that stray away from our center clustering, but none of the values are so far that they may be considered outliers for this particular test.

```python
# create visual of anova table and then of our plots of residuals #

vgmod = ols("Comfort ~ Material", data = vgcon).fit()
aov_table = sm.stats.anova_lm(vgmod, typ=1)
print(aov_table)

vg_res = vgmod.resid
fig = plt.figure()
ax = fig.add_subplot()

normality_plot, stat = stats.probplot(vg_res, plot=plt, rvalue= True)
ax.set_title("Probability plot of Grip Material residuals", fontsize= 20)
ax.set

plt.show()

# use shapiro test to determine independence #

print(stats.shapiro(vg_res))
```

Question 8. Belwo, you can see that the assumption of homogeneity is met both visually and using our BP test. Our residuals are all appearing in a straight line with only one type appearing to have an increased amount of variance. We can also see from our BP test that our p-value is greater than 0.05 meaning that we will fail to reject the null hypothesis of this test that states that we do not meet the homogeneity assumption.

```python
#creating a standardized plot#

stand_residuals = vgmod.get_influence().resid_studentized_internal

#fitted values#

fit_val = vgmod.fittedvalues

#create plot#

plt.figure(figsize=(8, 6))
plt.scatter(fit_val, stand_residuals)
plt.axhline(y=0, linestyle='dashed', color='red')
plt.axhline(y=3, color='blue')
plt.axhline(y=-3, color='blue')
plt.xlabel("Predicted y's")
plt.ylabel("Standardized Residuals")
plt.title("Standardized Residuals vs. Fitted Values")
plt.style.use('classic')
plt.show()

#run BP test by importing the function and then running the test#

from statsmodels.stats.diagnostic import het_breuschpagan

#Run BP test#

vgbp_test = het_breuschpagan(vgmod.resid,vgmod.model.exog)
f_statistic = vgbp_test[2]
f_p_value = vgbp_test[3]


print(f"F-statistic: {f_statistic}")
print(f"F p-value: {f_p_value}")
```

Question 9: Below, I ran the omnibus test to determine whether our null or alternative hypothesis is more strongly supported. Based on the omnibus test, we can see that we have an F (60,15) value that is equal to approximately 47.5. That F value is associated with a p value of less than 0.05. This indicates that the data more strongly supports our alternative hypothesis that there is one group that differs in comfort as rated by video game players substaintially from another.

```python
print(sm.stats.anova_lm(vgmod, typ=2))
```

```python
#Import Tukey function#

from statsmodels.stats.multicomp import pairwise_tukeyhsd

#Perform Tukey's HSD test#

tukey_result = pairwise_tukeyhsd(endog=vgcon['Comfort'], groups=vgcon['Material'],
                                 alpha=0.05)

print(tukey_result)
tukey_result.plot_simultaneous()
plt.show()
```

Question 10: Above, we ran Tukey's HSD post-hoc test to determine which materials' comfort levels are most different from each other. We can see that there is a significant difference between every group except for plastic and rubber. All of the p-adjusted values (with the exception of the plastic versus rubber) are also less than 0.05 which would indicate a stastistically signiciant difference between all of the groups except for the aforementioned two.
